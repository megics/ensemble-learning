```python
datasets = ['abalone','anneal','audiology','autos','balance-scale','breast-cancer','breast-w','col10','colic',
            'credit-a','credit-g','d159','diabetes','glass','heart-statlog','hepatitis','hypothyroid','ionosphere',
            'iris','kr-vs-kp','labor','letter','lymph','mushroom','primary-tumor','ringnorm','segment','sick',
            'sonar','soybean','splice','vehicle','vote','vowel','waveform','zoo']
```

```python
def read_data(filename):
    data,_ = arff.loadarff("36uci-dataset/"+filename+".arff")
    df = pd.DataFrame(data)
    return df
```

```python
dataframes = [0]*36
names = []

for i in range(0,36):
    dataframes[i] = read_data(datasets[i])
    names.append(datasets[i])
```

```python
def convert_cats(df):
    for n,c in df.items():
        if is_object_dtype(c):
            df[n] = c.astype("category").cat.as_ordered()
```

```python
def convert_int(df):
    col = [row for row in df]
    df[col[-1]] = df[col[-1]].cat.codes
```

```python
for i in range(0,36):
    convert_cats(dataframes[i])
    convert_int(dataframes[i])
```

```python
X_all = []
Y_all = []

for i in range(0,36):
    df = dataframes[i]
    x = df[df.columns[:-1]]
    y = df[df.columns[-1]]
    X_all.append(x.values)
    Y_all.append(y.values)
```

```python
text1
```

```python
def clean(text):
    wn = nltk.WordNetLemmatizer()
    stopword = nltk.corpus.stopwords.words('english')
    tokens = nltk.word_tokenize(text)
    lower = [word.lower() for word in tokens]
    no_stopwords = [word for word in lower if word not in stopword]
    no_alpha = [word for word in no_stopwords if word.isalpha()]
    lemm_text = [wn.lemmatize(word) for word in no_alpha]
    clean_text = lemm_text
    return clean_text
```

```python
def vectorize(data,tfidf_vect_fit):
    X_tfidf = tfidf_vect_fit.transform(data)
    words = tfidf_vect_fit.get_feature_names()
    X_tfidf_df = pd.DataFrame(X_tfidf.toarray())
    X_tfidf_df.columns = words
    return(X_tfidf_df)
```

```python
tfidf_vect = TfidfVectorizer(analyzer=clean)
tfidf_vect_fit = tfidf_vect.fit(X['text'])
X_text1 = vectorize(X['text'],tfidf_vect_fit)
```

```python
X_all.append(X_text1)
Y_all.append(Y_text1)
```

# 2) Splitting for all Data

```python
X_train_list = []
X_test_list = []
y_train_list = []
y_test_list = []
for i in range(0,38):
    X_train, X_test, y_train, y_test = train_test_split(X_all[i], Y_all[i], test_size=0.3, random_state=42) 
    X_train_list.append(X_train)
    X_test_list.append(X_test)
    y_train_list.append(y_train)
    y_test_list.append(y_test)
```

# 3) Base Learners

- Decision Tree (DT)
- Bagging (BG)
- Adaboost (AB)
- Random Forest (RF)
- Gradient Boosting (GBM)
- XGBoost (XGB)

```python
df_base = pd.DataFrame(columns=['Base'])
```

```python
for i in range(0,38):
    X_train = X_train_list[i]
    X_test = X_test_list[i]
    y_train = y_train_list[i]
    y_test = y_test_list[i]
    
    clf_base = DecisionTreeClassifier()
    clf_base.fit(X_train, y_train)
    score_base = clf_base.score(X_test, y_test)
    
    new_score = pd.Series({'Base':score_base}, name=names[i])    
    df_base = df_base.append(new_score, ignore_index=False)
```

# 4) Ensembles

- Ensemble methods with 150 trees.

```python
df_ensembles = pd.DataFrame(columns=['BG', 'AB', 'RF', 'GBM', 'XGB'])
```

```python
for i in range(0,38):
    #X_train, X_test, y_train, y_test = train_test_split(X_all[i].values, Y_all[i].values, test_size=0.3, random_state=42)
    
    X_train = X_train_list[i]
    X_test = X_test_list[i]
    y_train = y_train_list[i]
    y_test = y_test_list[i]
    
    clf_bg = BaggingClassifier(base_estimator=DecisionTreeClassifier(), n_estimators=150)
    clf_bg.fit(X_train, y_train)
    score_bg = clf_bg.score(X_test, y_test)
    
    clf_ab = AdaBoostClassifier(base_estimator=DecisionTreeClassifier(),n_estimators=150)
    clf_ab.fit(X_train, y_train)
    score_ab = clf_ab.score(X_test, y_test)

    clf_rf = RandomForestClassifier(n_estimators=150)
    clf_rf.fit(X_train, y_train)
    score_rf = clf_rf.score(X_test, y_test)
    
    clf_gbm = GradientBoostingClassifier(n_estimators=150) 
    clf_gbm.fit(X_train, y_train)
    score_gbm = clf_gbm.score(X_test, y_test)
    
    clf_xgb = XGBClassifier(n_estimators=150, use_label_encoder=False, eval_metric='logloss')
    clf_xgb.fit(X_train, y_train)
    score_xgb = clf_xgb.score(X_test, y_test)
    
    new_score = pd.Series({'BG':score_bg, 'AB':score_ab, 'RF':score_rf, 'GBM':score_gbm, 'XGB':score_xgb}, name=names[i])    
    df_ensembles = df_ensembles.append(new_score, ignore_index=False)
```

# 5) Ensemble in Ensembles (Deep Ensembles)

## 5.1) RF + GBM + XGB (hard voting)

```python
result1 = []

estimators1 = []

clf_rf1 = RandomForestClassifier(n_estimators=50)
clf_gbm1 = GradientBoostingClassifier(n_estimators=50) 
clf_xgb1 = XGBClassifier(n_estimators=50, use_label_encoder=False, eval_metric='logloss')

estimators1.append(('rf', clf_rf1))
estimators1.append(('gbm', clf_gbm1))
estimators1.append(('xgb', clf_xgb1))

model_hard1 = VotingClassifier(estimators1)

for i in range(0,38):
    X_train = X_train_list[i]
    X_test = X_test_list[i]
    y_train = y_train_list[i]
    y_test = y_test_list[i]
    #X_train, X_test, y_train, y_test = train_test_split(X_all[i].values, Y_all[i].values, test_size=0.3, random_state=42)
          
    model_hard1.fit(X_train, y_train)
    score_1 = model_hard1.score(X_test, y_test)
    
    result1.append(score_1)
```
